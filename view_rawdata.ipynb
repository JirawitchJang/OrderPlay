# Clean data

missing values


```python
df.info()
df.head()
```

```python
missing_values = df.isna().sum()
missing_columns = missing_values[missing_values > 0]

print(missing_columns)
```

```python
column_names = list(missing_columns.index)

print("Missing Columns\n")

i = 1

for c in column_names:
    print(f"[{i}]",c)
    i += 1
    
df[missing_columns.index].head()
```

```python
df['Instructions'] = df['Instructions'].fillna("No Instructions")
df['Discount construct'] = df['Discount construct'].fillna("No Discount construct")
# df['Rating'] = df['Rating'].fillna(0)
df['Review'] = df['Review'].fillna("No Review")
df['Cancellation / Rejection reason'] = df['Cancellation / Rejection reason'].fillna('No reason')
df['Restaurant compensation (Cancellation)'] = df['Restaurant compensation (Cancellation)'].fillna('No Cancellation')
df['Restaurant penalty (Rejection)'] = df['Restaurant penalty (Rejection)'].fillna('No rejection')
# df['KPT duration (minutes)'] = df['KPT duration (minutes)'].fillna(0)
# df['Rider wait time (minutes)'] = df['Rider wait time (minutes)'].fillna(0)
df['Customer complaint tag'] = df['Customer complaint tag'].fillna("None")

df[missing_columns.index].head()
```

For NaN column (Number)

```python
df_raw = pd.read_csv("order_history_kaggle_data.csv")
df_fn = df_raw.copy()

print("0 คือค่าที่เป็น False มีค่าอยู่ไม่ NaN \n1 คือค่าที่เป็น True คือในที่นี้คือค่านี้เป็น NaN")
 
# เพื่อเอามาเเยกว่าข้อมูลตัวแปรที่เป็น int มีลักษณะยังไง เพื่อดูความสัมพันธ์กัน

df_fn['KPT_missing'] = df_raw['KPT duration (minutes)'].isna().replace({True:'Not Record',False:'Recorded'})
df_fn['Rating_missing'] = df_raw['Rating'].isna().replace({True:'Unrated',False:'Rated'})
df_fn['Rider_missing'] = df_raw['Rider wait time (minutes)'].isna().replace({True:'Not Record',False:'Recorded'})

df_fn['Rating'] = df['Rating'].fillna(0)
df_fn['KPT duration (minutes)'] = df['KPT duration (minutes)'].fillna(0)
df_fn['Rider wait time (minutes)'] = df['Rider wait time (minutes)'].fillna(0)

df_fn[['KPT duration (minutes)','KPT_missing','Rating','Rating_missing','Rider wait time (minutes)','Rider_missing']]

```

```python
df['Instructions'] = df['Instructions'].fillna("No Instructions")
df['Discount construct'] = df['Discount construct'].fillna("No Discount construct")
df['Rating'] = df['Rating'].fillna(0)
df['Review'] = df['Review'].fillna("No Review")
df['Cancellation / Rejection reason'] = df['Cancellation / Rejection reason'].fillna('No reason')
df['Restaurant compensation (Cancellation)'] = df['Restaurant compensation (Cancellation)'].fillna('No Cancellation')
df['Restaurant penalty (Rejection)'] = df['Restaurant penalty (Rejection)'].fillna('No rejection')
df['KPT duration (minutes)'] = df['KPT duration (minutes)'].fillna(0)
df['Rider wait time (minutes)'] = df['Rider wait time (minutes)'].fillna(0)
df['Customer complaint tag'] = df['Customer complaint tag'].fillna("None")

df[missing_columns.index].head()
```
